**Question 1:** write a function `isValid(s)` that takes as argument a string s containing a sequence of parenthesis '(', ')', '{', '}', '[' and ']', and determines if the input is valid. A input string is valid if for every open parenthensis there is a close one and parenthesis is well-formed. e.g "(){}[]" is valid.


```python
def isValid(s: str) -> bool:

  stack = []
  mapping = {")": "(", "}": "{", "]": "["}
  for char in s:
    if char in mapping:
      if not stack or stack.pop() != mapping[char]:
        return False
    else:
      stack.append(char)

  return not stack


print(isValid("(){}[]")) 
print(isValid("([)]"))  
print(isValid("("))  
print(isValid(")"))  
    

...


```

**Question 3:** Write an infinite generator function `odd_squares_sum` that yields the sum of square of odd numbers. e.g $1^2 + 3^2 + 5^2 + ...$ up to a ``limit``

```python
def odd_squares_sum(limit):
    current = 1
    while current ** 2 <= limit:
        yield current ** 2
        current += 2

# Testing the generator
theGen = odd_squares_sum(12)
print(next(theGen))  # Should print 1
print(next(theGen))  # Should print 9

# Continuing the test case to verify the function's correctness
theGen = odd_squares_sum(50)
assert next(theGen) == 1
assert next(theGen) == 9
assert next(theGen) == 25
assert next(theGen) == 49

...
```

```python
def greatest_common_divisor(a, b):
    
    while b != 0:
        a, b = b, a % b
    return a


print(greatest_common_divisor(10, 15)) 
print(greatest_common_divisor(21, 14))  
print(greatest_common_divisor(100, 25)) 

...
```

**Question 7:**  Write a function `get_3_nearest` that takes in a point of interest ``pt`` and a **list** of points ``ptlist``  and returns a list of 3 nearest points from the point of interest ``pt``. Assume the distance between any two point is defined by the `L1-norm`.

```python
def get_l1_distance(pt1, pt2):
    """
    Compute the L1-norm (Manhattan distance) between two points pt1 and pt2.
    """
    return sum(abs(a - b) for a, b in zip(pt1, pt2))

def get_3_nearest(pt, ptlist):
    """
    Find the 3 nearest points in ptlist to the point pt using L1-norm (Manhattan distance).
    """
    # Calculate distances from pt to each point in ptlist
    distances = [(get_l1_distance(pt, p), p) for p in ptlist]
    
    # Sort distances by distance (first element of tuple)
    distances.sort(key=lambda x: x[0])
    
    # Extract the 3 nearest points
    nearest_points = [p for _, p in distances[:3]]
    
    return nearest_points


pt = (0, 0)
ptlist = [(1, 1), (2, 2), (3, 3), (0, 1), (5, 5), (-1, -1)]
nearest = get_3_nearest(pt, ptlist)
print("3 Nearest points to", pt, "are:", nearest)

    


...
```

Example: given `flatten_reverse_lists([[2,13,44], [6,7]])` it should return `[2,6,7,13,44]`

```python
def flatten_reverse_lists(list_of_lists):
    """
    Flatten the list of lists and return a reverse-sorted list of elements.
    
    Parameters:
    list_of_lists (list): List containing sublists.
    
    Returns:
    list: Reverse-sorted list of elements from all sublists.
    """
    # Flatten the list of lists into a single list
    flattened_list = [item for sublist in list_of_lists for item in sublist]
    
    # Sort the flattened list in reverse order
    sorted_list = sorted(flattened_list, reverse=True)
    
    return sorted_list

# Example usage:
result = flatten_reverse_lists([[2, 13, 44], [6, 7]])
print(result)  # Output: [44, 13, 7, 6, 2]

    

...
```

**Question 10:** Do the following to the dataframe:
* Create a new collumn ``total_price`` whose value is equal to $scoops * price$*
* Write a function ``groupStatistics(data, groupValue)``. Internally, this function groups ``data``  by ``flavor`` and then returns statistics of a given grouped item ``groupValue`` indexed on the ``total_price`` columns. The statistics is a numpy array contains ``[mean, media, min, max, std]`` of the ``total_price`` column. The ``std`` should be rounded to 2 **decimal places**



```python
import pandas as pd
import numpy as np

# Creating the DataFrame
data = pd.DataFrame({
    'flavor': ['white chocolate', 'vanilla', 'dark chocolate', 'strawberry', 'strawberry', 
               'vanilla', 'mint', 'mint', 'white chocolate', 'dark chocolate', 
               'white chocolate', 'dark chocolate'],
    'scoops': [1, 1, 2, 1, 3, 2, 1, 2, 3, 3, 2, 5],
    'price': [2, 1.5, 3, 2, 4, 2, 4, 5, 2, 3, 2, 3]
})

# Adding a new column 'total_price'
data['total_price'] = data['scoops'] * data['price']

def groupStatistics(data, groupValue):
    # Group by 'flavor' and calculate statistics on 'total_price'
    group_data = data.groupby('flavor')['total_price'].describe()
    
    # Extracting statistics for the given groupValue
    stats = group_data.loc[groupValue, ['mean', '50%', 'min', 'std']]
    
    # Rounding 'std' to 2 decimal places
    stats['std'] = round(stats['std'], 2)
    
    # Converting to numpy array
    stats_array = np.array(stats)
    
    return stats_array


flavor = 'mint'
statistics = groupStatistics(data, flavor)
print(f"Statistics for '{flavor}' flavor:")
print(f"Mean: {statistics[0]}, Median: {statistics[1]}, Min: {statistics[2]}, Std: {statistics[3]}")

...
```
